```python
import pandas as pd
```

```python
dados = pd.read_csv('./aluguel.csv', sep=';')
```

```python
residencial_tipo = ['Quitinete', 'Casa', 'Apartamento', 'Casa de Condomínio', 'Casa de Vila', 'Chácara',]
residencial_tipo
```

```python
dados_residenciais = dados[dados.Tipo.isin(residencial_tipo)]
dados_residenciais
```

```python
dados_apartamento = dados.Tipo == 'Apartamento'
dados[dados_apartamento].Tipo.count()
```

```python
dados_casa_condominio_ou_vila = dados.Tipo.isin(['Casa de Condomínio','Casa de Vila'])
dados[dados_casa_condominio_ou_vila]
```

```python
dados_60_100_m2 = dados.Area.between(60, 100)
dados[dados_60_100_m2]
```

```python
dados_4_quartos_aluguel_menor_2000 = (dados.Quartos >= 4) & (dados.Valor < 2000)
dados[dados_4_quartos_aluguel_menor_2000]
```
